```python
import urllib.request as u
from langdetect import detect_langs
from  bs4 import BeautifulSoup

p=[]
z=[]
#wiki = "https://www.lokmat.com/pune/free-bus-service-savitribai-phule-pune-university-today/"
links = ["https://www.matrubharti.com/book/read/content/19866266/pathlag-1/","https://www.matrubharti.com/book/read/content/19866381/pathlag-2","https://www.matrubharti.com/book/read/content/19866496/pathlag-3","https://www.matrubharti.com/book/read/content/19866551/pathlag-4","https://www.matrubharti.com/book/read/content/19866610/pathlag-5", "https://www.matrubharti.com/book/read/content/19866701/pathlag-6", "https://www.matrubharti.com/book/read/content/19866759/pathlag-7"]

def extract_data(wiki):
    page = u.urlopen(wiki)
    soup = BeautifulSoup(page, "lxml")
    # # Selecting by 'a' Tag and printing all hyperlinks observed
    
    for link in soup.find_all('a', href=True):
        #print(link['href'])
        p.append(link['href'])
    # # Selecting all 'p' tags and printing their text 
    all_tables2 = soup.find_all('p')
    # # Select only Marathi Part by using langdetect

    for i in all_tables2 :
        i = i.getText()
        try:
            res = detect_langs(i)
        except:
            pass
        for item in res:
            if item.lang == "mr" :
                if(len(i.split())>=5):
                    if(len(i.split()) >= 13):
                        str=i
                        for j in str.split('.'):
                            if(len(j.split()) >= 5 ):
                                if j not in z:
                                    if(len(z) < 1000):
                                        z.append(j)
                                        print(j)
                                    else:
                                        break
                                        
                    else:
                        if i not in z:
                            if(len(z) < 1000):
                                z.append(i)
                                print(i)
                            else:
                                break
    #print(z)
    print(len(z))

for i in links:
    extract_data(i)
with open("test.txt",'wb') as f:
    for i in range(0,1000):
        f.write(("<s>"+z[i]+"</s>\n").encode('utf-8'))
```
